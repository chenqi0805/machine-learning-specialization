```python
import graphlab
```

```python
products = graphlab.SFrame('amazon_baby.gl/')
```

```python
products.head()
```

```python
selected_words=['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products['word_count']
```

```python
def selected_count(i,x):
    if i in x:
        y=x[i]
    else:
        y=0
    return y
```

```python
for i in selected_words:
    products[i]=products['word_count'].apply(lambda x: selected_count(i,x))
```

```python
count=[0 for i in range(11)]
```

```python
for i in range(11):
    count[i]=products[selected_words[i]].sum()
```

## Out of the selected_words, which one is most used in the dataset? Which one is least used?

```python
count
```

```python
#ignore all 3* reviews
products = products[products['rating'] != 3]
```

```python
#positive sentiment = 4* or 5* reviews
products['sentiment'] = products['rating'] >=4
```

```python
len(products['sentiment'])
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=selected_words,
                                                     validation_set=test_data)
```

```python
selected_words_model['coefficients']
```

## Out of the 11 words in selected_words, which one got the most positive weight? Which one got the most negative weight? 

```python
selected_words_model['coefficients'].sort('value').tail()
```

```python
selected_words_model.evaluate(test_data)
```

```python
selected_words_model.evaluate(test_data, metric='roc_curve')
```
